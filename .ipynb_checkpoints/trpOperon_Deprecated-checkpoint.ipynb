```python
import numpy as np
from numpy.linalg import eig, det
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
sns.set_palette('pastel')
import sympy as sp
from scipy.integrate import odeint
from scipy.optimize import fsolve
```

### Define Differential Equations

```python
def trp_eqs(v, u_R):
    """
    Define parameter and ODE's describing the trp operon
    x = x[0] : mRNA conc
    y = x[1] : enzyme conc
    z = x[2] : trp conc
    """
    x = v[0]
    y = v[1]
    z = v[2]
    
    # Constants
    a_1 = 0.9
    a_2 = 0.02
    a_3 = 0.0
    a_4 = 0.024
    a_5 = 430.0
    a_6 = -7.5
    k = 0.005
    k_i = 100.0
    T = 0.0156
    mu_m = 0.0312
    
    k_a = 5.5
    k_r = 25.0
    r_m = 20.0
    
    r = r_m * z / (z + k_r)
    mu = u_R * mu_m
    u = T * mu
    
    # ODE's
    dx_dt = (1.0 + z) / (1.0 + (1.0 + r) * z) * k_a / (k_a + z) - (u + a_1) * x
    dy_dt = x - (u + a_2) * y
    dz_dt = y * (k_i**2) / ((k_i**2) + (z**2)) -\
    (u + a_3) * z - (a_4 * z) / (1 + z) * z / (z + k_r) -\
    a_5 * (1.0 + a_6 * u) * u * z / (z + k)
    
    return [dx_dt, dy_dt, dz_dt]
```

```python
def trp_dynamics(v, t, u_R):
    """
    Helper function for simulating trp operon dynamics over time
    """
    return trp_eqs(v, u_R)
```

### Solve for the Fixed Points
By solving for all DE's = 0 numerically

```python
def u_R_to_ss_concs(u_R_vec):
    x_stars = []
    y_stars = []
    z_stars = []
    for u_R in u_R_vec:
        x_star, y_star, z_star = fsolve(trp_eqs, (1.0, 1.0, 1.0), args=(u_R,))
        x_stars.append(x_star)
        y_stars.append(y_star)
        z_stars.append(z_star)
    x_stars = np.array(x_stars)
    y_stars = np.array(y_stars)
    z_stars = np.array(z_stars)
    return x_stars, y_stars, z_stars
```

```python
u_R_vals = np.arange(0.0, 1.0, 0.01)
```

```python
x_stars, y_stars, z_stars = u_R_to_ss_concs(u_R_vals)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))
ax1.plot(u_R_vals, y_stars, 'r', label='Enzyme')
ax1.plot(u_R_vals, z_stars, 'g', label='Trp')
ax1.set_xlabel('relative growth rate')
ax1.set_ylabel('Enzyme & Trp Conc')

ax2 = ax1.twinx()
ax2.plot(u_R_vals, x_stars, 'b', label='mRNA')
ax2.set_ylabel('mRNA Conc')
fig.tight_layout()
```

```python
x0 = [x_star, y_star, z_star]
t = np.linspace(0, 500, 1000)
```

### Jacobian Matrix

```python
(x, y) = sp.symbols('x, y', real=True)
```

```python
dx, dy = trp_eqs((x, y))
```

```python
A = sp.Matrix([dx, dy])
```

```python
J = A.jacobian([x, y])
```

```python
J
```

```python
J_atFP = J.subs([(x, x_star), (y, y_star)])
J_atFP
```

### Evaluating Stability
